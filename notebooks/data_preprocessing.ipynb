4. Cleaning the data

```python
#Created a copy of the df to make changes in for a clean version without affecting the original raw data
c_df = df.copy()

#Displaying the new df to validate it was copied correctly
c_df.head()
```

```python
#Based on what was found before, the 'Context' column was removed entirely
c_df = c_df.drop(columns=['Context'])
```

```python
#Displayed to validate the column was removed successfully
c_df.head()
```

```python
#Change null values to dsiplay something else
c_df = c_df.fillna(value = {'Longitude':'No Location',
                     'Latitude': 'No Location',
                     'LSOA code': 'No location',
                     'LSOA name': 'No location',
                    'Crime ID': 'N/A',
                    'Last outcome category': 'N/A'})
```

```python
#Validate there are no nulls
c_df.isnull().sum()
```

```python
#Removed the duplicate rows
c_df = c_df.drop_duplicates()
```

```python
#Verify there are no longer any duplicates
c_df.duplicated().sum()
```

```python
#Change data type of values in 'Month' column from object to date
df["Month"] = pd.to_datetime(df["Month"])
```
